```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/Learning_ML/Linear Regression/homeprices.csv")
df
```

```python
plt.scatter(df.area, df.price, c='red', marker='+')
plt.xlabel('area(sqr ft)')
plt.ylabel('price($)')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df[['price']])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(df.area, df.price, c='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), c='blue')
plt.xlabel('area(sqr ft)', fontsize=15)
plt.ylabel('price($)', fontsize=15)
```

```python
newdf = pd.read_csv('/content/drive/MyDrive/Learning_ML/Linear Regression/areas.csv')
newdf.head()
```

```python
p = reg.predict(newdf)
```

```python
newdf['price'] = p
newdf.head()
```

```python
newdf.to_csv("/content/drive/MyDrive/Learning_ML/Linear Regression/prediction.csv", index=False)
```
